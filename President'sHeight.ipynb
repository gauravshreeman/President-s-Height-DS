President's Height 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("heights.csv")
data
```

```python
height = np.array(data["height(cm)"])  //heights in array form 
height
```

```python
name = np.array(data["name"]) //names in array form
name
```

```python
data.head()
```

```python
data.tail()
```

```python
print("Mean of Heights: ", height.mean());
print("Standard Deviation: ", height.std());
print("Minnimum Height: ", height.min());
print("Maximum Height: ", height.max());
```

```python
print("25th Percentile: ", np.percentile(height, 25))
print("Median: ", np.median(height))
print("95th Percentile: ", np.percentile(height, 95))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.hist(height)
plt.title("Height Distribution of presidents")
plt.xlabel("height")
plt.ylabel("order")
plt.show()
```

```python
plt.plot(height)
plt.title("Height Distribution of presidents")
plt.xlabel("height")
plt.ylabel("order")
plt.show()
```

```python
plt.fill_between(height, name)
plt.title("Height Distribution of presidents")
plt.xlabel("Height")
plt.ylabel("Name")
plt.show()
```
